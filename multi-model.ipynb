## Custom NER just for understanding

```python
import spacy
from spacy.training import Example
import random

# 1. Load blank English model
nlp = spacy.blank("en")

# 2. Create NER pipeline
if "ner" not in nlp.pipe_names:
    ner = nlp.add_pipe("ner")
else:
    ner = nlp.get_pipe("ner")

# 3. Add your custom labels
labels = ["BRAND", "PRODUCT"]  # Add whatever labels you need
for label in labels:
    ner.add_label(label)

# 4. Training data (custom)
TRAIN_DATA = [
    ("Apple just released a new iPhone", {"entities": [(0, 5, "BRAND"), (21, 27, "PRODUCT")]}),
    ("Samsung announces Galaxy S22", {"entities": [(0, 7, "BRAND"), (17, 25, "PRODUCT")]}),
    ("Google launches Pixel 7", {"entities": [(0, 6, "BRAND"), (15, 21, "PRODUCT")]}),
]

# 5. Start training
optimizer = nlp.begin_training()
for i in range(30):  # 30 epochs, adjust for more data
    random.shuffle(TRAIN_DATA)
    losses = {}
    for text, annotations in TRAIN_DATA:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        nlp.update([example], drop=0.2, losses=losses)
    print(f"Epoch {i}, Losses: {losses}")
```

```python

# 6. Test the trained model
test_text = "got cool pixel7"
doc = nlp(test_text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```
